This dataset is data of the top 2000 most popular songs on Spotify (as of three years ago when the dataset was published) from Kaggle. The data includes popularity of each song as well as other aspects like beats per minute and loudness.
We chose this dataset because we thought it would be interesting to examine the listening patterns of listeners around the world, and to see what (if any) correlations there were between different elements of each song.
Hopefully, we can gain some understanding of what makes a song popular, if there is any thread connecting the most popular songs, or at least gain some knowledge about all of the different songs in the dataset. 

### Numpy

```python
# numpy: https://numpy.org/doc/stable/user/absolute_beginners.html
 

BPMmean = (df["Beats Per Minute (BPM)"].mean())
BPMstd = (df["Beats Per Minute (BPM)"].std())
BPMmed = (df["Beats Per Minute (BPM)"].median())


Energymean = (df["Energy"].mean())
Energystd = (df["Energy"].std())
Energymed = (df["Energy"].median())

Dancemean = (df["Danceability"].mean())
Dancestd = (df["Danceability"].std())
Dancemed = (df["Danceability"].median())

Loudmean = (df["Loudness (dB)"].mean())
Loudstd = (df["Loudness (dB)"].std())
Loudmed = (df["Loudness (dB)"].median())

Valmean = (df["Valence"].mean())
Valstd = (df["Valence"].std())
Valmed = (df["Valence"].median())

Speechmean = (df["Speechiness"].mean())
Speechstd = (df["Speechiness"].std())
Speechmed = (df["Speechiness"].median())

numpy_array = np.array([[BPMmean, Energymean, Dancemean, Loudmean, Valmean, Speechmean],
                        [BPMstd, Energystd, Energystd, Loudstd, Valstd, Speechstd], 
                        [BPMmed, Energymed, Dancemed, Loudmed, Valmed, Speechstd]])

import pandas as pd

# Create the dataframe
df = pd.DataFrame(numpy_array,
                  columns=['Beats Per Minute (BPM)',
                           'Energy',
                           'Danceability',
                           'Loudness (dB)',
                           'Valence',
                           'Speechiness'],
                  index=['mean', 'median', 'mode'])

df.head()



# Calculate the following statistics on a select 6 feeatures of your dataset

# mean calculation

# standard deviation calculation

# median calculation

# organize the calculated statistics into a dataframe that displays the mean, standard deviation, and median per each feature
```
